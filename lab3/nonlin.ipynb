```python
def norm_vec(vec):
    return np.linalg.norm(vec, ord=np.inf)

def mat_norm(matrix):
    return np.linalg.norm(matrix, ord=np.inf)

threshold = 1e-10

def check_res(A, u, f, eps=threshold):
    norm = norm_vec(np.matmul(A, u) - f)
    return norm, abs(norm) < eps
```

```python
# Метод простой итерации
def fpi(phi, deriv_phi, x0,  l_lim, r_lim, max_iter=100, eps=threshold, ref_root=None):
    e = np.min([np.abs(x0 - l_lim), np.abs(x0 - r_lim)])
    if np.abs(deriv_phi(e)) >= 1:
        print("phi derivative must be less than one to converge", np.abs(deriv_phi(e)))
        return None, None
        
    x = x0
    diff = []

    for i in range(max_iter):
        if not (l_lim < x and x < r_lim):
            print("Out of limits", x)
            return None, None
        old_x = x
        x = phi(x)

        d = np.abs(x - old_x)
        diff.append(d)

        if d < eps:
            break

    return x, diff
```

```python
# Метод Ньютона
def Newton(phi, deriv_phi, x0,  l_lim, r_lim, max_iter=100, eps=threshold, ref_root=None):        
    x = x0
    diff = []

    for i in range(max_iter):
        if not (l_lim < x and x < r_lim):
            print("Out of limits", x)
            return None, None
        old_x = x
        x = old_x - phi(old_x) / deriv_phi(x)

        d = np.abs(x - old_x)
        diff.append(d)

        if d < eps:
            break

    return x, diff
```

```python
def partial(f, x, var_index, h):
    dim = len(x)

    vec_h = np.zeros(dim)
    vec_h[var_index] = h

    return (f(x+vec_h) - f(x-vec_h)) / (2*h)

def getAbsJacobi(f, x, h):
    dim = len(f(x))
    J = np.zeros((dim, dim))

    for i in range(dim):
        for j in range(dim):
            J[i][j] = partial(lambda x: f(x)[i], x, j, h[i])
    
    return J
```

```python
# Метод простой итерации
def fpiSyst(phi, x0,  l_lim, r_lim, max_iter=100, eps=threshold, ref_root=None):
    dim = len(phi(x0))
    h = 1e-3 * np.array([np.min([np.abs(x0[i] - l_lim[i]), np.abs(x0[i] - r_lim[i])]) for i in range(dim)])
    x = x0

    J = getAbsJacobi(phi, x0, h)
    
    sum_J = np.zeros(dim)
    for i in range(dim):
        for j in range(len(J[i])):
            sum_J[i] += J[i][j]

    max_sum_J = np.max(sum_J)
    print(sum_J)
    if max_sum_J > 1:
        print("jacobi colums sum must be less than one to converge")
        return None, None
        
    diff = []
    for i in range(max_iter):
        if not (np.all(l_lim < x) and np.all(x < r_lim)):
            print("Out of limits", x)
            return None, None
        old_x = x
        x = phi(x)
        d = np.abs(x - old_x)
        d_norm = np.linalg.norm(d, ord=np.inf)
        diff.append(d_norm)
        if d_norm < eps:
            break

    return x, diff
```

```python
def Newton_Sys(phi, x0, l_lim, r_lim, max_iter=100, eps=threshold, ref_root=None):
    dim = len(phi(x0))
    h = 1e-3 * np.array([np.min([np.abs(x0[i] - l_lim[i]), np.abs(x0[i] - r_lim[i])]) for i in range(dim)])
    x = x0

    diff = []
    for i in range(max_iter):        
        if not (np.all(l_lim < x) and np.all(x < r_lim)):
            print("Out of limits", x)
            return None, None

        old_x = x
        x = x - np.matmul(np.linalg.inv(getAbsJacobi(phi, x, h)), phi(x))
        d = np.abs(x - old_x)
        h = d
        d_norm = np.linalg.norm(d, ord=np.inf)
        diff.append(d_norm)
        
        if d_norm < eps:
            break

    return x, diff
```
