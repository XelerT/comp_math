### Задача:

предъявить численное решение уравнения переноса и сравнить его с аналитическим (постановку аналитической задачи найти самостоятельно): 
 - использовать схемы для численного решения: уголок, прямоугольник, Лакс-Вендрофф (Явная четырехточечная схема)
- провести исследование на сходимость (сравнение численного и аналитического решения)
- провести анализ порядка аппроксимации (сходимости) (график с убыванием нормы ошибки в зависимости от шага интегрирования в логарифмическом масштабе)

```python
def solve_transport_sequential(X: float, T: float, M: int, K: int, a: float,
                               phi: Callable[[float], float],
                               psi: Callable[[float], float],
                               f_source: Callable[[float, float], float],
                               comm=None):
    u = np.zeros((K + 1, M + 1))
    h = X / M
    tau = T / K

    for m_idx in range(M + 1):
        u[0, m_idx] = phi(m_idx * h)

    for k_idx in range(K):
        u[k_idx + 1, 0] = psi((k_idx + 1) * tau)

        for m_idx in range(1, M):
            term_avg_space = 0.5 * (u[k_idx, m_idx + 1] + u[k_idx, m_idx - 1])
            term_transport = (tau * a / (2 * h)) * (u[k_idx, m_idx + 1] - u[k_idx, m_idx - 1])
            term_source = tau * f_source(k_idx * tau, m_idx * h)
            u[k_idx + 1, m_idx] = term_avg_space - term_transport + term_source

        if M > 1:
            u[k_idx + 1, M] = 2 * u[k_idx + 1, M - 1] - u[k_idx + 1, M - 2]
        elif M == 1:
             u[k_idx + 1, M] = u[k_idx + 1, 0]
    return u
```

```python
def mms_convergence_study(
    X: float, T: float, M: int, K: int, a: float,
    u_exact_func: Callable[[float, float], float],
    solver_func: Callable = solve_transport_sequential
) -> tuple[float, float]:
    h = X / M
    tau = T / K

    delta_t = 1e-8 
    delta_x = 1e-8

    du_dt_lambda = lambda t_val, x_val: (u_exact_func(t_val + delta_t, x_val) - u_exact_func(t_val - delta_t, x_val)) / (2 * delta_t)
    du_dx_lambda = lambda t_val, x_val: (u_exact_func(t_val, x_val + delta_x) - u_exact_func(t_val, x_val - delta_x)) / (2 * delta_x)

    f_mms_lambda = lambda t_val, x_val: du_dt_lambda(t_val, x_val) + a * du_dx_lambda(t_val, x_val)
    phi_mms_lambda = lambda x_val: u_exact_func(0, x_val)
    psi_mms_lambda = lambda t_val: u_exact_func(t_val, 0)

    u_numerical = solver_func(X, T, M, K, a, phi_mms_lambda, psi_mms_lambda, f_mms_lambda)

    u_exact_final_t = np.zeros(M + 1)
    for m_idx in range(M + 1):
        u_exact_final_t[m_idx] = u_exact_func(T, m_idx * h)

    u_numerical_final_t = u_numerical[K, :]

    abs_error_vector = np.abs(u_numerical_final_t - u_exact_final_t)

    norm_u_exact = np.linalg.norm(u_exact_final_t)
    if norm_u_exact < 1e-15:
        l2_error = np.linalg.norm(u_numerical_final_t - u_exact_final_t)
    else:
        l2_error = np.linalg.norm(u_numerical_final_t - u_exact_final_t) / norm_u_exact

    linf_error = np.max(abs_error_vector)

    return l2_error, linf_error
```

```python
def analyze_approximation_order_spatial(
    X: float, T: float, a: float,
    u_exact_func: Callable[[float, float], float],
    M_values: list[int],
    solver_func: Callable = solve_transport_sequential,
    cfl_target: float = 0.5
):
    h_values = []
    l2_errors = []
    linf_errors = []

    for M_val in M_values:
        h = X / M_val
        if a == 0:
            K_val = M_val * 2
        else:
            K_val = math.ceil((T * abs(a)) / (cfl_target * h))

        tau = T / K_val
        actual_cfl = abs(a) * tau / h if h > 0 else float('inf')

        if actual_cfl > 1.0 and a !=0:
            print(f"Warning: M={M_val}, K={K_val} -> CFL = {actual_cfl:.2f} > 1. Might be unstable.")

        h_values.append(h)
        l2_err, linf_err = mms_convergence_study(X, T, M_val, K_val, a, u_exact_func, solver_func)
        l2_errors.append(l2_err)
        linf_errors.append(linf_err)

    h_values = np.array(h_values)
    l2_errors = np.array(l2_errors)
    linf_errors = np.array(linf_errors)

    if len(h_values) >= 2:
        if l2_errors[-1] > 0 and l2_errors[-2] > 0 :
            p_l2 = (math.log(l2_errors[-1]) - math.log(l2_errors[-2])) / \
                   (math.log(h_values[-1]) - math.log(h_values[-2]))
            print(f"\nEstimated L2 convergence order (last two points): {p_l2:.2f}")
        else:
            print("error, ", l2_errors[-1] > 0, l2_errors[-2] > 0)

        if linf_errors[-1] > 0 and linf_errors[-2] > 0:
            p_linf = (math.log(linf_errors[-1]) - math.log(linf_errors[-2])) / \
                     (math.log(h_values[-1]) - math.log(h_values[-2]))
            print(f"Estimated Linf convergence order (last two points): {p_linf:.2f}")
        else:
            print("Could not estimate Linf convergence order due to non-positive errors.")


    plt.figure(figsize=(10, 6))
    plt.loglog(h_values, l2_errors, 'o-', label='L2 Error')
    plt.loglog(h_values, linf_errors, 's--', label='L-infinity Error')

    if len(h_values) > 0 and l2_errors[0]>0 : # Check l2_errors[0] to avoid issues with C1, C2
        C1 = l2_errors[0] / h_values[0]
        plt.loglog(h_values, C1 * h_values, 'k:', label='O(h)')
        C2 = l2_errors[0] / (h_values[0]**2)
        plt.loglog(h_values, C2 * (h_values**2), 'k-.', label='O(h^2)')

    plt.xlabel('Step size h (log scale)')
    plt.ylabel('Error Norm (log scale)')
    plt.title('Convergence Rate Analysis (Spatial)')
    plt.grid(True, which="both", ls="-")
    plt.legend()
    plt.gca().invert_xaxis()
    plt.savefig("convergence_plot_spatial_jupyter.png")
    print("\nConvergence plot saved to convergence_plot_spatial_jupyter.png")
    plt.show()
```

```python
X_param = 2.0 * np.pi
T_param = 1.0
a_param = 1.0

def u_exact(t: float, x: float) -> float:
    return np.sin(x - a_param * t) + np.cos(x + a_param * t)

print("--- Performing Single MMS Convergence Study ---")
M_single = 100
h_single = X_param / M_single
cfl_val = 0.5
if a_param == 0:
    K_single = M_single * 2
else:
    K_single = math.ceil((T_param * abs(a_param)) / (cfl_val * h_single))

l2_err_single, linf_err_single = mms_convergence_study(
    X_param, T_param, M_single, K_single, a_param, u_exact
)
print(f"For M={M_single}, K={K_single}:")
print(f"  L2 Error: {l2_err_single:.4e}")
print(f"  Linf Error: {linf_err_single:.4e}")

M_values_for_study = [20, 40, 80, 160, 320] # Example values, ensure they are not too small for K calculation

analyze_approximation_order_spatial(
    X_param, T_param, a_param, u_exact, M_values_for_study, cfl_target=cfl_val
)
```
