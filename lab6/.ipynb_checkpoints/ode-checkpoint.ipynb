## Полезные вещи

```python
def partial(f, x, var_index, h):
    dim = len(x)
    assert var_index < dim, "Index of variable is greater than dimensity."

    vec_h = np.zeros(dim)
    vec_h[var_index] = h

    return (f(x+vec_h) - f(x-vec_h)) / (2*h)

def getJacobi(f, x, h):
    dim = len(f(x))
    assert dim == len(x), "f and x must be of same size."
    J = np.zeros((dim, dim))

    for i in range(dim):
        for j in range(dim):
            J[i][j] = partial(lambda x: f(x)[i], x, j, h[i])

    return J

def solveSystemNewton(f, x0, n_iter, precision, bl, tr):
    assert precision > 0, "Precision must be positive."
    dim = len(f(x0))
    assert len(x0) == dim and len(bl) == dim and len(tr) == dim, "f, x0, bl and tr must be of same size."
    h = [np.min([np.abs(x0[i] - bl[i]), np.abs(x0[i] - tr[i])]) for i in range(dim)]
    x = x0

    diff = []
    for i in range(n_iter):
        assert np.all(bl < x) and np.all(x < tr), f"Gone out of bounds on iter {i}. Diff: {diff}."
        old_x = x
        x = x - np.matmul(np.linalg.inv(getJacobi(f, x, h)), f(x))
        d = np.abs(x - old_x)
        h = d
        d_norm = np.linalg.norm(d, ord=np.inf)
        diff.append(d_norm)
        if d_norm < precision:
            break

    return x, diff
```

### Задача:

Решить системы ОДУ следующими классами методов:

### Исходные данные:

```python
t_begin = 0
t_end = 100
y0 = np.array([2, 0])
def f(t, vec, e):
    x, z = vec
    result = [
        z,
        e * (1 - x**2) * z - x
    ]

    return np.array(result)
```

```python
def drawResult(res_t, res_y, e, h):
    x, z = res_y.transpose()

    fig, axes = plt.subplots(2, figsize=(10, 10))

    fig.suptitle(f'e={e}, h={h}')
    
    axes[0].scatter(res_t, x, label='x', s=5)
    axes[0].scatter(res_t, z, label='z', s=5)

    axes[1].scatter([0], [0], label='center', s=20, color='red')
    axes[1].scatter(x, z, label='trajectory', s=5)

    xlim = axes[1].get_xlim()
    axes[1].set_xlim(xlim[0]*1.2, xlim[1]*1.2)
    
    axes[1].legend()
    axes[0].legend()
```

## Рунге-Кутты

```python
def classicRK4(t0, y0, h, f, n_iter=None, t_end=None, e=None):
    if n_iter is None:
        n_iter = math.ceil((t_end - t_begin) / h)
    if e is None:
        e = 0.1

    t = t0
    y = y0
    result_t = [t]
    result_y = [y]

    for i in range(n_iter):
        k1 = f(t        , y, e)
        k2 = f(t + 1/2*h, y + 1/2*k1*h, e)
        k3 = f(t + 1/2*h, y + 1/2*k2*h, e)
        k4 = f(t +     h, y +     k3*h, e)
        y = y + h*(1/6*k1 + 1/3*k2 + 1/3*k3 + 1/6*k4)

        t += h
        result_t.append(t)
        result_y.append(y)

    return np.array(result_t), np.array(result_y)
```

```python
result_t, result_y = classicRK4(t_begin, y0, 0.1, f, t_end=t_end, e=0.1)
drawResult(result_t, result_y, 0.1, 0.1)

result_t, result_y = classicRK4(t_begin, y0, 0.1, f, t_end=t_end, e=0.5)
drawResult(result_t, result_y, 0.5, 0.1)
```

## Метод Адамса

```python
def multistepAdams3(t0, y0, h, f, n_iter=None, t_end=None, e=None):
    if n_iter is None:
        n_iter = math.ceil((t_end - t_begin) / h)
    if e is None:
        e = 0.1

    result_t, result_y = classicRK4( t0, y0, h, f, n_iter=2, e=e)
    result_t = list(result_t)
    result_y = list(result_y)
    saved_f  = []
    for i in range(3):
        saved_f.append(f(result_t[i], result_y[i], e))

    t = result_t[2]
    y = result_y[2]
    for i in range(2, n_iter):
        y = y + h*(23/12*saved_f[i] - 16/12*saved_f[i-1] + 5/12*saved_f[i-2])

        t += h
        result_t.append(t)
        result_y.append(y)
        saved_f.append(f(t, y, e))

    return np.array(result_t), np.array(result_y)
```

```python
result_t, result_y = multistepAdams3(t_begin, y0, 0.1, f, t_end=t_end, e=0.1)
drawResult(result_t, result_y, 0.1, 0.1)
```

```python
result_t, result_y = multistepAdams3(t_begin, y0, 0.1, f, t_end=t_end, e=0.5)
drawResult(result_t, result_y, 0.5, 0.1)
```

```python
def BDF3(t0, y0, h, f, n_iter=None, t_end=None, e=None):
    if n_iter is None:
        n_iter = math.ceil((t_end - t_begin) / h)
    if e is None:
        e = 0.1

    result_t, result_y = classicRK4(t0, y0, h, f, n_iter=2, e=e)

    deviation = np.max(2*np.add.accumulate(np.abs(result_y))[-1])

    result_t = list(result_t)
    result_y = list(result_y)

    t = result_t[2]
    y = result_y[2]
    for i in range(2, n_iter):
        func = lambda arg: arg - 18/11*result_y[i] + 9/11*result_y[i-1] - 2/11*result_y[i-2] - h*6/11*f(t, arg, e)
        y, _ = solveSystemNewton(func, y, 10, 1e-4, y-deviation, y+deviation)

        y_ref = op.fsolve(func, y)
        assert np.linalg.norm(y_ref - y, ord=np.inf) < 1e-4

        t += h
        result_t.append(t)
        result_y.append(y)

    return np.array(result_t), np.array(result_y)
```

```python
result_t, result_y = BDF3(t_begin, y0, 0.1, f, t_end=t_end, e=0.1)
drawResult(result_t, result_y, 0.1, 0.1)
```

```python
result_t, result_y = BDF3(t_begin, y0, 0.1, f, t_end=t_end, e=0.5)
drawResult(result_t, result_y, 0.5, 0.1)
```
