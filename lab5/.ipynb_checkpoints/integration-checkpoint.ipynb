```python
# Метод трапеции

def int_trapec(h, x, f):
    n = int(abs(x[0] - x[len(x)-1]) / h)
    
    sum = (f[0] + f[n]) / 2
    for i in range(1, n):
        sum += f[i]

    return h * sum

def trapezoidal_rule(x, f_x):
    h = x[1] - x[0]  # Interval width (assumes uniform spacing)
    integral = h * (0.5 * f_x[0] + 0.5 * f_x[len(f_x)-1] + np.sum(f_x[1:len(f_x)-1]))
    return integral
```

```python
h = np.average(np.diff(data['x']))
print('Результат методом трапеции: ', int_trapec(h, data['x'], data['f']))
print('Referance', np.trapezoid(x=data['x'], y=data['f']))
```

```python
# уточнение результата правилом Рунге

def int_trapec_Runge(h, x, f):
    order=2
    I_h = int_trapec(h, x, f)
    
    return I_h + (I_h - int_trapec(2*h, x, f)) / (2**order - 1)
```

```python
h = np.average(np.diff(data['x']))
print('Результат методом трапеции с уточнением по правилу Рунге: ', int_trapec_Runge(h, data['x'], data['f']))
```

```python
# Метод Симпсона

def int_Simpson(h, x, f):
    n = int(abs(x[0] - x[len(x)-1]) / h)
    
    sum = f[0] + f[n]
    sub_sum = 0
    
    for i in range(1, int(n/2) + 1):
        sub_sum += f[2*i - 1]

    sum += 4 * sub_sum
    sub_sum = 0
    
    for i in range(1, int(n/2)):
        sub_sum += f[2*i]

    sum += 2 * sub_sum

    return h * sum / 3
```

```python
h = np.average(np.diff(data['x']))
print('Результат методом Симпсона: ', int_Simpson(h, data['x'], data['f']))
print('Referance', sp.integrate.simpson(x=data['x'], y=data['f']))
```

```python
def cyclic_thomas(n, f, lower, diag, upper):
    cmod = np.zeros(n)
    v = np.zeros(n)

    # Forward elimination
    cmod[1] = upper[1] / diag[1]
    v[1] = -lower[1] / diag[1]
    f[1] /= diag[1]

    for i in range(2, n - 1):
        m_factor = 1.0 / (diag[i] - lower[i] * cmod[i - 1])
        cmod[i] = upper[i] * m_factor
        v[i] = -lower[i] * v[i - 1] * m_factor
        f[i] = (f[i] - lower[i] * f[i - 1]) * m_factor

    # Handle last row
    m_last = 1.0 / (diag[n - 1] - lower[n - 1] * cmod[n - 2])
    cmod[n - 1] = upper[n - 1] * m_last
    v[n - 1] = (-upper[0] - lower[n - 1] * v[n - 2]) * m_last
    f[n - 1] = (f[n - 1] - lower[n - 1] * f[n - 2]) * m_last

    # Backward substitution
    for i in range(n - 2, 0, -1):
        v[i] -= cmod[i] * v[i + 1]
        f[i] -= cmod[i] * f[i + 1]

    # Solve for m[0]
    f[0] = (f[0] - lower[0] * f[n - 1] - upper[0] * f[1]) / (
        diag[0] + lower[0] * v[n - 1] + upper[0] * v[1]
    )

    f[1:] += f[0] * v

    return f


def cubic_spline_interpolation(h, b):
    n = len(h)  # Number of intervals
    
    # Construct the cyclic tridiagonal system
    diag = 2 * (h[:-1] + h[1:])  # Main diagonal
    lower = h[1:-1]              # Lower diagonal
    upper = h[1:-1]              # Upper diagonal
    f = 6 * (b[1:] - b[:-1])   # Right-hand side

    diag = np.hstack([1, diag, 1])  # Correct main diagonal size
    f = np.hstack([0, f, 0])    # Correct f size
    lower = np.hstack([0, h[:-1]])  # Pad lower diagonal
    upper = np.hstack([h[:-1], 0])  # Pad upper diagonal

    return cyclic_thomas(lower.shape[0], f, lower, diag, upper)
    

def cubic_polynomial(x, y):

    n = len(x) - 1  # Number of intervals
    h = np.diff(x)  # Spacing between consecutive x values
    b = np.diff(y) / h  # Slope between consecutive points
    m = cubic_spline_interpolation(h, b)
        
    splines = []

    def poly(arg):
        int_ind = np.searchsorted(x, arg)
        int_ind = np.clip(int_ind, 1, n)
        
        val = m[int_ind-1] * (x[int_ind] - arg)**3 / (6 * h[int_ind-1])                                   \
              + m[int_ind] * (arg - x[int_ind-1])**3 / (6 * h[int_ind-1])                                 \
              + (y[int_ind-1] - m[int_ind-1] * h[int_ind-1]**2 / 6) * (x[int_ind] - arg) / h[int_ind-1]   \
              + (y[int_ind] - m[int_ind] * h[int_ind-1]**2 / 6) * (arg - x[int_ind-1]) / h[int_ind-1]
        return val

    return lambda arg: poly(arg)
```

Для интеграла, с весовой функцией $sin(2nx)$ значения коэффициентов и x равны:

$$
-x_1 = x_2 = \sqrt{\pi^2 - \frac{3}{2n^2}}
$$

$$
c_1 = -c_2 = \frac{\pi}{2n \sqrt{\pi^2 - \frac{3}{2n^2}}}
$$

$$
I_G = c_1 f(x_1) + c_2 f(x_2)
$$

При $sin(kx), k=30$ => $n = k / 2$

```python
from scipy.integrate import quad

x2 = np.sqrt(np.pi**2 - 6 / k**2)
x1 = -x2

c1 = np.pi / (k * np.sqrt(np.pi**2 - 6 / k**2))
c2 = - c1

print(x1, x2, c1, c2)

Ig = c1 * equ(x1) + c2 * equ(x2)
print('I_G=', Ig)
print('Referance quad', quad(equ, min(x), max(x))[0])
print('Referance Трапеция', np.trapezoid(x=x, y=y))
print('Referance Симпсон', sp.integrate.simpson(x=x, y=y))
```
